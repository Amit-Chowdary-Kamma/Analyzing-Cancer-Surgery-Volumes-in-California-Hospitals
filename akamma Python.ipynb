```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Amit\Desktop\AIT PJ\Cancer_Surgeries_Volume_California.csv", delimiter=';')
surgery_counts = df.groupby('Surgery')['# of Cases'].sum().reset_index()
surgery_counts = surgery_counts.sort_values(by='# of Cases', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(surgery_counts['Surgery'], surgery_counts['# of Cases'], color='purple', edgecolor='yellow')
plt.xlabel('Surgery Type')
plt.ylabel('Total Number of Surgeries')
plt.title('Total Number of Surgeries Performed for Each Surgery Type')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\Amit\Desktop\AIT PJ\Cancer_Surgeries_Volume_California.csv", delimiter=';', skiprows=range(1, 101))
df_yearly = df.groupby('Year')['# of Cases'].sum().reset_index()

plt.figure(figsize=(9, 5))
plt.plot(df_yearly['Year'], df_yearly['# of Cases'], marker='o', linestyle='-', color='blue', 
         markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=1.5)
plt.title('Trend of Cancer Surgeries Volume Over the Years (All Types)')
plt.xlabel('Year')
plt.ylabel('Number of Surgeries')
plt.grid(True)
plt.xticks(df_yearly['Year'], rotation=45)  
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt


county_boundaries = gpd.read_file("C:/Users/Amit/Downloads/ca_county_boundaries.shp")
surgery_data = pd.read_csv("C:/Users/Amit/Downloads/cancervolume1.csv", encoding='latin1')
merged_data = county_boundaries.merge(surgery_data, left_on='NAME', right_on='County', how='left')


fig, ax = plt.subplots(1, 1, figsize=(10, 10))
merged_data.plot(column='number_of_cases', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Surgery Volume by County')
plt.show()
```
